```python
#Body mass index
weight=float(input("Enter the BMI weight:"))
if(weight<=18):
    print("underweight")
elif(weight<=18.5):
    print("thin for height")
elif(weight>=18.6 and weight<=24.9):
    print("healthy weight")
elif(weight>=25 and weight<=29.9):
    print("overweight")
else:
    print("Very Overweight")
```
